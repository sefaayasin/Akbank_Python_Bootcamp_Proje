```python
import csv
import datetime
```

```python
class Pizza:
    def __init__(self, description, size):
        self.description = description
        self.size = size
        
        if self.size == 1:
            self.cost = 89
        elif self.size == 2:
            self.cost = 99
        elif self.size == 3:
            self.cost = 109
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
```

```python
class Margarita(Pizza):
    
    def __init__(self, size):
        super().__init__("Margarita", size)

    def get_description(self):
        return "Margarita / pizza sosu, mozzarella peyniri ve domates"
    
class Sucuklu(Pizza):
    
    def __init__(self, size):
        super().__init__("Sucuklu", size)

    def get_description(self):
        return "Sucuklu / pizza sosu, mozzarella peyniri, dilim zeytin ve yeşil biber"

class Salamlı(Pizza):
    
    def __init__(self, size):
        super().__init__("Salamlı", size)
    
    def get_description(self):
        return "Salamlı / pizza sosu, mozzarella peyniri, salam, dilim zeytin ve yeşil biber,"
    
class Sosisli(Pizza):
    
    def __init__(self, size):
        super().__init__("Sosisli", size)
    
    def get_description(self):
        return "Sosisli / pizza sosu, mozzarella peyniri, sosis, dilim zeytin ve yeşil biber,"
    
class Mantarlı(Pizza):
    
    def __init__(self, size):
        super().__init__("Mantarlı", size)
    
    def get_description(self):
        return "Mantarlı / pizza sosu, mozzarella peyniri, mantar, dilim zeytin ve yeşil biber,"

class Akdeniz(Pizza):
    
    def __init__(self, size):
        super().__init__("Akdeniz", size)
    
    def get_description(self):
        return "Akdeniz / pizza sosu, mozzarella peyniri, hellim, beyaz peynir, köz patlıcan, yeşil biber, dilim zeytin ve soğan,"
    
class Egeli(Pizza):
    
    def __init__(self, size):
        super().__init__("Egeli", size)
    
    def get_description(self):
        return "Egeli / pizza sosu, mozzarella peyniri, beyaz peynir, mantar, domates, mısır, zeytin, yeşil biber,"
    
class Chicken(Pizza):
    
    def __init__(self, size):
        super().__init__("Chicken", size)
    
    def get_description(self):
        return "Chicken / pizza sosu, mozzarella peyniri, tavuk parçacıkları, mısır, zeytin ve yeşil biber,"

class Kumru(Pizza):
    
    def __init__(self, size):
        super().__init__("Kumru", size)
    
    def get_description(self):
        return "Kumru / pizza sosu, mozzarella peyniri, julyen doğranmış sucuk&salam, mısır, yeşil biber,"
    
class Mangal_Zamanı(Pizza):
    
    def __init__(self, size):
        super().__init__("Mangal Zamanı", size)
    
    def get_description(self):
        return "Mangal Zamanı / pizza sosu, mozarella peyniri, misket köfte, küp sucuk, mantar, köz kırmızı biber, meksika biberi, soğan,"
    
class İtalyan_Aşkı(Pizza):
    
    def __init__(self, size):
        super().__init__("İtalyan Aşkı", size)
    
    def get_description(self):
        return "İtalyan Aşkı / pizza sosu, mozzarella peyniri, küp sucuk, mantar, mısır, zeytin ve kırmızı köz biber,"

class Karışık(Pizza):
    
    def __init__(self, size):
        super().__init__("Karışık", size)
    
    def get_description(self):
        return "Karışık / pizza sosu, mozzarella peyniri, sucuk, salam, sosis, mantar, zeytin ve yeşil biber,"

class Meksikalı(Pizza):
    
    def __init__(self, size):
        super().__init__("Meksikalı", size)
    
    def get_description(self):
        return "Meksikalı / pizza sosu, mozzarella peyniri, kıyma, mısır, meksika biberi, soğan,"
    
class Peynirsever(Pizza):
    
    def __init__(self, size):
        super().__init__("Peynirsever", size)
    
    def get_description(self):
        return "Peynirsever / pizza sosu, mozzarella peyniri, beyaz peynir, hellim peyniri, cheddar peyniri,"

class Yakamoz(Pizza):
    
    def __init__(self, size):
        super().__init__("Yakamoz", size)
    
    def get_description(self):
        return "Yakamoz / pizza sosu, mozzarella peyniri, ton balığı, mısır, soğan, yeşil biber,"
   
```

```python
class Decorator(Pizza):
    def __init__(self, pizza, description, cost):
        super().__init__(pizza.get_description() + ", " + description, pizza.size)
        self.pizza = pizza
        self._description = description
        self._cost = cost

    def get_description(self):
        return self._description

    def get_cost(self):
        return self._cost
```
